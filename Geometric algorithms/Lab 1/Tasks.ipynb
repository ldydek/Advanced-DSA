```python
def det1(a, b, c):
    return ((a[0]*b[1]*1) + (b[0]*c[1]*1) + (c[0]*a[1]*1)) - ((c[0]*b[1]*1) + (b[0]*a[1]*1) + (a[0]*c[1]*1))
def det2(a, b, c):
    return ((a[0]-c[0]) * (b[1] - c[1])) - ((b[0] - c[0]) * (a[1] - c[1]))
def det1np(a, b, c):
    matrix = np.array([[a[0], a[1], 1], [b[0], b[1], 1], [c[0], c[1], 1]])
    return np.linalg.det(matrix)
def det2np(a, b, c):
    matrix = np.array([[a[0]-c[0], a[1]-c[1]], [b[0]-c[0], b[1]-c[1]]])
    return np.linalg.det(matrix)
```

```python
a = [-1.0, 0.0]
b = [1.0, 0.1]
s = []
epsilon = [10**(-8), 10**(-10), 10**(-12), 10**(-14)]
det = [0, 1, 2, 3]
for x in range(len(epsilon)):
    for y in range(len(det)):
        s.append("eps "+str(epsilon[x])+", det "+str(det[y]))
epsilon_s = ["10**(-8)", "10**(-10)", "10**(-12)", "10**(-14)"]
det_s = ["0", "1", "2", "3"]
```

```python
def classpoint(a, b, c, det, eps):
    if det == 0:
        determinant = det1(a, b, c)
    elif det == 1:
        determinant = det2(a, b, c)
    elif det == 2: 
        determinant = det1np(a, b, c)
    elif det == 3:
        determinant = det2np(a, b, c)
    
    if determinant > eps:
        return -1
    elif determinant < -eps:
        return 1
    else:
        return 0
    
# -1 - po lewej stronie prostej
# 1 - po prawej stronie prostej
# 0 - na prostej
```
