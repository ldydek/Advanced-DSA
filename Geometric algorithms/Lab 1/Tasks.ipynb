# Rozwiązania

```python
import numpy as np
from matplotlib import pyplot as plt
```

# zadanie 1

# a)

```python
points_a = [(0, 0) for _ in range(10**5)]
for x in range(10**5):
    points_a[x] = (np.random.uniform(-10**4, 10**4), np.random.uniform(-10**5, 10**5))
for x in range(10):
    print(points_a[x])
```

# b)

```python
points_b = [(0, 0) for _ in range(10**5)]
for x in range(10**5):
    points_b[x] = (np.random.uniform(-10**14, 10**14), np.random.uniform(-10**14, 10**14))
for x in range(10):
    print(points_b[x])
```

# c)

```python
points_c = [(0, 0) for _ in range(10**3)]
for x in range(10**3):
    angle = np.random.uniform(0, 2*np.pi)
    points_c[x] = (100*np.cos(angle), 100*np.sin(angle))
for x in range(10):
    print(points_c[x])
```

# d)

```python
a = [-1.0, 0.0]
b = [1.0, 0.1]
# wyliczam współczynnik kierunkowy oraz wyraz wolny
# y = ax + b
# 1) 0 = -a + b
# 2) 0.1 = a + b
# więc 0.1 = a + b = 2a
# a = 0.05
# b = 0.05
points_d = [(0, 0) for _ in range(10**3)]
for x in range(10**3):
    a = np.random.uniform(1, 10**3)
    points_d[x] = (a, 0.05*a+0.05)
for x in range(10):
    print(points_d[x])
```
