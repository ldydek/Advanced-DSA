```python
x = np.linspace(0, 2, 100)
plt.xlabel('Oś odciętych')
plt.ylabel('Oś rzędnych')
plt.title('y = x^2')
plt.plot(x, x**2)
```

```python
fig = plt.figure()
x = np.linspace(-100, 100, 1000)
plt.plot(x, x**3, "r")
plt.axis([-10, 10, -30, 30])
plt.xlabel('Oś odciętych')
plt.ylabel('Oś rzędnych')
plt.title('y = x^3')
plt.show()
```
